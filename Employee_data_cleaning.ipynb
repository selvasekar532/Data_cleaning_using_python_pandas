```python
duplicates = df[df.duplicated()]
```

```python
# Droping all the duplicate rows
```

```python
df.drop_duplicates(inplace = True)
```

```python
# Recheck the column count after drop all the duplicate rows
```

```python
df.info()
```

```python
print(df.to_string())
```

```python
# Finding the most repeadted name...
```

```python
most = df['Name'].value_counts().idxmax()
```

```python
most
```

```python
# Replacing all the null values with the most repeated value
```

```python
df['Name'].fillna(most,inplace = True)
```

```python
# Printing all the values to see everthing gets changed or not!
```

```python
print(df.to_string())
```

```python
# Change the one numerical value to most repeated value
```

```python
df['Name'].replace(12345, most, inplace = True)
```

```python
# Checking the changes has been made!
```

```python
print(df.to_string())
```

```python
df.info()
```
